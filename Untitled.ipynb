```python
def folds(m,train, label, test, num_sp, rnd_st):
    prediction = pd.DataFrame(np.zeros(shape = (test.shape[0], num_sp*16)))
    prediction.columns = [a for a in range(num_sp*16)]
    testing = pd.DataFrame(np.zeros(shape = (test.shape[0], 16)))
    testing.columns = [a for a in range(16)]
    a=[]
    b=[]
    splitter= StratifiedKFold(n_splits = num_sp ,shuffle=True, random_state = rnd_st)
    
    for i , (tr_index,te_index) in enumerate(splitter.split(train,label)):
        x_train,x_val = train.loc[tr_index,:], train.loc[te_index,:]
        y_train, y_val = label[tr_index], label[te_index]
        
        m.fit(x_train,y_train)
        tr_pred = m.predict_proba(x_train)
        val_pred = m.predict_proba(x_val)
     #   return(m.predict_proba(test))
        prediction.loc[:, 16*i:16*i+15] = m.predict_proba(test)
        
        print(f'training : {log_loss(pd.get_dummies(y_train).values,tr_pred)}  \
              testing : {log_loss(pd.get_dummies(y_val).values, val_pred)}')
        
        a.append(log_loss(pd.get_dummies(y_train).values,tr_pred))
        b.append(log_loss(pd.get_dummies(y_val).values, val_pred))
    
    print(f'training mean : {np.mean(a)}        testing mean : {np.mean(b)}')
    
    for i in range(16):
        testing.loc[:,i] = (prediction.loc[:,i]+prediction.loc[:,i+16]+prediction.loc[:,i+32])/3
    
    return(testing)
```
